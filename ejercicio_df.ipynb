importar libreria

```python
import pandas as pd
```

crear un dataframe

```python
motos={
    "marca": ["ducatti","akt","yamaha","BMW"],
    "cilindraje":[600,150,1000,1200],
    "modelo":[2015,2018,2026,2022],
}

df=pd.DataFrame(motos)
display(df)
```

```python
print(df["marca"][2])
```

modificar un dato

```python
df.loc[1,"marca"]="honda"
df["modelo"]=[1999,1995,1993,1998]
display(df)
```

modificar un valor

```python
df.loc[1,"cilindraje"]=500
display(df)
```

Modificar una columna

```python
df["color"]=["negro","azul","verde","rojo"]
display(df)
```

```python
df.loc[4]=["suzuki",800,2000,"gris"]
display(df)
```

como poner una fila 

```python
df.loc[len(df)]=["kawasaki",1000,2003,"amarilla"]
display(df)
```

filtrar un registro

```python
df_filtrado=df["cilindraje"]==1000
display(df_filtrado)
```

filtrado estructura nombre = df [condicion]

```python
df_filtrado=df[df["cilindraje"]== 1000]
display(df_filtrado)
```

```python
df_filtrado=df[df["modelo"]>1998]
display(df_filtrado)
```
