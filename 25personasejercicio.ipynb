```python
import pandas as pd
```

```python
df_personas=pd.read_csv(r"C:\Users\User\Documents\Ejercicio de clase - Hoja 1.csv",header=1)
display(df_personas)
```

```python
df_personas=df_personas.drop(columns=["Unnamed: 0", "Unnamed: 1" ])
display(df_personas)
```

```python
df_personas.rename(columns={"Nombre":"NOMBRE","Apellido":"APELLIDO","Edad":"EDAD","Género":"GENERO", "Nivel educativo":"NIVEL_EDUCATIVO"}, inplace=True)
display(df_personas)
```

```python
df_personas=df_personas.dropna(subset=["NOMBRE","APELLIDO"])
display(df_personas)
```

```python
df_personas.info()
```

```python
df_personas.loc[9,"EDAD"]= None
df_personas.loc[18,"EDAD"]= None
df_personas.loc[19,"EDAD"]= None
df_personas.loc[25,"EDAD"]= None
df_personas.loc[36,"EDAD"]= None
df_personas
```

```python
df_personas["EDAD"]=df_personas["EDAD"].astype("Float64")
df_personas["EDAD"]=df_personas["EDAD"].astype("Int64")
df_personas.info()
```

```python
prom=((df_personas["EDAD"]).mean()).astype(int)
prom
```

```python
df_personas["GENERO"]=df_personas["GENERO"].replace({"M":"Masculino","F":"Femenino","m":"Masculino","f":"Femenino","femenino":"Femenino","Indefinido":None})
df_personas
```

```python
df_Femeninos= df_personas[df_personas["GENERO"] == "Femenino"]
display(df_Femeninos)
cantidad_Feminino=df_Femeninos["GENERO"].count()
print("La cantidad de mujeres es:", cantidad_Feminino)
```

```python
df_Masculinos= df_personas[df_personas["GENERO"] == "Masculino"]
display(df_Masculinos)
cantidad_Masculino=df_Masculinos["GENERO"].count()
print("La cantidad de mujeres es:", cantidad_Masculino)
```

```python
data= df_personas.groupby("GENERO")["NOMBRE"].count()
print(data)
```

```python
df_personas["NOMBRE"]=df_personas["NOMBRE"].str.capitalize()
df_personas
```

```python
df_personas["APELLIDO"]=df_personas["APELLIDO"].str.capitalize()
df_personas
```

```python
data1= df_personas.groupby("NIVEL_EDUCATIVO")["NOMBRE"].count()
print(data1.sort_values(ascending=False))
```

```python
df_personas["NIVEL_EDUCATIVO"]=df_personas["NIVEL_EDUCATIVO"].str.capitalize()
data1= df_personas.groupby("NIVEL_EDUCATIVO")["NOMBRE"].count()
print(data1.sort_values(ascending=False))
```

```python
df_personas.loc[df_personas["NIVEL_EDUCATIVO"].str.contains("Uni"), "NIVEL_EDUCATIVO"]="Universidad"
df_personas.loc[df_personas["NIVEL_EDUCATIVO"].str.contains("Bach"), "NIVEL_EDUCATIVO"]="Bachillerato"
df_personas.loc[df_personas["NIVEL_EDUCATIVO"].str.contains("Sec"), "NIVEL_EDUCATIVO"]="Bachillerato"
df_personas.loc[df_personas["NIVEL_EDUCATIVO"].str.contains("Prof"), "NIVEL_EDUCATIVO"]="Universidad"
df_personas
```
