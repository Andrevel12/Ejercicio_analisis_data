```python
import pandas as pd

list_columnas=["producto","precio","cantidad"]
list_data=[
    ["Manazanas",3000,10],
    ["Peras",2500,5],
    ["Bananas",2000,8],
    ["Mandarina",3000,12],
    ["Naranjas",2000,20],
    ["Fresas",2500,7],
]
df_frutas=pd.DataFrame(data=list_data,columns=list_columnas)
display(df_frutas)
```

agregar columnas, multiplicar dos columnas

```python
df_frutas["valor_total"]= df_frutas["precio"]*df_frutas["cantidad"]
display(df_frutas)
```

filtrado

```python
df_filtrado=df_frutas[df_frutas["valor_total"]>20000]
display(df_filtrado)
```

productos con mayores ventas

```python
mayor_venta=df_frutas["valor_total"].max()
producto_max=df_frutas[df_frutas["valor_total"]== mayor_venta]
display(producto_max["producto"])
```

el producto con mayor ventas

```python
pos_max = df_frutas["valor_total"].idxmax()
display(df_frutas.loc[pos_max,"producto"])
```

descuento del 10%

```python
df_frutas["descuento"]=(df_frutas["valor_total"]-df_frutas["valor_total"]*0.1).astype(int)
display(df_frutas)
```

ventas totales

```python
venta_totales= df_frutas["valor_total"].sum()
print("ventas totales", venta_totales)
```

```python
prom_totales=(df_frutas["valor_total"].mean()).astype(int)
print("promedio ventas totales",prom_totales)
```

```python
df_ordenado=df_frutas.sort_values("valor_total", ascending=False)
display(df_ordenado)
```
